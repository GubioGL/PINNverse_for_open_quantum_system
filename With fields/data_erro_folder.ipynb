```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.stats import pearsonr

import random
import torch as tc
```

```python
for N in [5,10,15,20,25,50]:
    # Loop para carregar os arquivos e calcular os erros
    for i in range(1, 101):
        caminho = f"/home/gubio/CODESPACE/Pinn-inverse-for-opem-quantum-system/With fields/raw_data/parametro_withfields_N{N}_seed{i}_std0.csv"
        df = pd.read_csv(caminho, index_col=0)
        
        # Calculando erros
        valor_real = df['treino'].str.strip('[]').astype(float)
        valor_previsto = df['previsto'].str.strip('[]').astype(float)
        
        erro_abs = abs(valor_real - valor_previsto)# Absolute Error
        erro_mape = abs(valor_real - valor_previsto)/abs(valor_real) # Mean Absolute Percentage Error
        erro_mse = (valor_real - valor_previsto)**2# Mean Squared Error

        

        df['MAPE'] = erro_mape
        df['MSE'] = erro_mse
        df['AE'] = erro_abs

        # Salvando os resultados em um novo arquivo CSV
        novo_caminho = f"/home/gubio/CODESPACE/Pinn-inverse-for-opem-quantum-system/With fields/data_erro/erros_parametro_withfields_N{N}_seed{i}_std0.csv"
        df.to_csv(novo_caminho)
```
