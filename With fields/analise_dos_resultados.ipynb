```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.stats import pearsonr

import random
import torch as tc
```

```python
for N in [5,10,15,20,25,50]:
    # Loop para carregar os arquivos e calcular os erros
    for i in range(5, 101):

        np.random.seed(i)
        tc.manual_seed(i) 
        random.seed(i)
        ################## Definindo os parametros #######################
        Js          = [random.uniform(-1,1) for _ in range(15)]
        dissipation = [random.uniform(0,1) for _ in range(4)]
        parametro_treino = {
            'gamma1': [dissipation[0]],
            'gamma2': [dissipation[1]],
            'gamma3': [dissipation[2]],
            'gamma4': [dissipation[3]],
            'JIX': [Js[0]],
            'JIY': [Js[1]],
            'JIZ': [Js[2]],
            'JXI': [Js[3]],
            'JYI': [Js[4]],
            'JZI': [Js[5]],
            'JXX': [Js[6]],
            'JXY': [Js[7]],
            'JXZ': [Js[8]],
            'JYX': [Js[9]],
            'JYY': [Js[10]],
            'JYZ': [Js[11]],
            'JZX': [Js[12]],
            'JZY': [Js[13]],
            'JZZ': [Js[14]]
            }

        caminho = f"C:/Users/gubio/Pesquisa/PINN for open quantum system/With fields/data/parametro_withfields_N{N}_seed{i}_std0.csv"
        caminho_new = f"C:/Users/gubio/Pesquisa/PINN for open quantum system/With fields/data2/parametro_withfields_N{N}_seed{i}_std0.csv"

        df = pd.read_csv(caminho, index_col=0)

        for key, value in parametro_treino.items():
            df.at[key, 'treino'] = value
        df.to_csv(caminho_new)
```

```python
for N in [5,10,15,20,25,50]:
    # Inicializando listas para armazenar os erros acumulados
    erros_abs_acumulados = []

    # Loop para carregar os arquivos e calcular os erros
    for i in range(5, 101):
        caminho = f"C:/Users/gubio/Pesquisa/PINN for open quantum system/With fields/data2/parametro_withfields_N{N}_seed{i}_std0.csv"
        df = pd.read_csv(caminho, index_col=0)
        
        # Calculando erros
        valor_real = df['treino'].str.strip('[]').astype(float)
        valor_previsto = df['previsto'].str.strip('[]').astype(float)
        
        erro_abs = np.abs(valor_real - valor_previsto)/abs(valor_real) # MAPE
        
        # Removendo Js0 e Js1
        #erro_abs = erro_abs.drop(index=['JIX', 'JIY','JIZ','JXI','JYI','JZI'], errors='ignore')
        
        # Acumulando erros
        erros_abs_acumulados.append(erro_abs)

    # Calculando a média, desvio padrão, máximo e mínimo dos erros
    erros_abs_concat = pd.concat(erros_abs_acumulados, axis=1)
    
    tabela_erros = pd.DataFrame({
        "MAPE": abs(erros_abs_concat.mean(axis=1)),
        "Desvio Padrão": erros_abs_concat.std(axis=1),
        "Máximo": erros_abs_concat.max(axis=1),
        "Mínimo": erros_abs_concat.min(axis=1)
    })

    print(tabela_erros)
    
    # Calculando as barras de erro
    yerr = [tabela_erros['MAPE'] - tabela_erros['Mínimo'], 
        tabela_erros['Máximo'] - tabela_erros['MAPE']]

    # Plotando o gráfico de pontos com barras de erro em escala logarítmica
    #plt.figure(figsize=(10, 6))
    plt.title("N = " + str(N))
    names = [r'$\gamma_1$', r'$\gamma_2$', r'$\gamma_3$', r'$\gamma_4$',
             r'$JXX$',r'$JXY$',r'$JXZ$',r'$JYX$',r'$JYY$',r'$JYZ$',r'$JZX$',
             r'$JZY$',r'$JZZ$',r'$JIX$',r'$JIY$',r'$JIZ$',r'$JXI$',r'$JYI$',r'$JZI$']
    plt.errorbar(names, tabela_erros['MAPE'], yerr=yerr, fmt='o', capsize=5)
    plt.yscale('log')
    plt.xlabel('Parâmetros', fontsize=14)
    plt.ylabel('Valores', fontsize=14)
    plt.yticks(fontsize=14)
    #plt.ylim(1e-7, 1e-3)  # Definindo os limites do eixo y
    plt.grid(True)
    plt.show()
```

```python
# Inicializando listas para armazenar os erros acumulados para cada parâmetro
erros_abs_acumulados_por_parametro = {j: [] for j in ['gamma1', 'gamma2', 'gamma3', 'gamma4']}

for N in [5,10,15,20,25,50]:
    # Loop para carregar os arquivos e calcular os erros
    for i in range(5, 101):
        caminho = f"C:/Users/gubio/Pesquisa/PINN for open quantum system/With fields/data2/parametro_withfields_N{N}_seed{i}_std0.csv"
        df = pd.read_csv(caminho, index_col=0)
        
        # Calculando erros
        valor_real = df['treino'].str.strip('[]').astype(float)
        valor_previsto = df['previsto'].str.strip('[]').astype(float)
        
        erro_abs = np.abs(valor_real - valor_previsto)/valor_real # MAPE
        
        # Removendo Js0 e Js1
        erro_abs = erro_abs.drop(index=['JIX', 'JIY','JIZ','JXI','JYI','JZI'], errors='ignore')
        
        # Acumulando erros por parâmetro
        for j in erros_abs_acumulados_por_parametro.keys():
            if j in erro_abs.index:
                erros_abs_acumulados_por_parametro[j].append((N, erro_abs[j]))

fig, axs = plt.subplots(1, 4, figsize=(25, 6), sharey=True)

for ax, (j, valores) in zip(axs, erros_abs_acumulados_por_parametro.items()):
    valores_df = pd.DataFrame(valores, columns=['N', 'Erro'])
    
    # Calculando a média, mínimo e máximo
    media = valores_df.groupby('N')['Erro'].mean()
    minimo = valores_df.groupby('N')['Erro'].min()
    maximo = valores_df.groupby('N')['Erro'].max()
    
    # Calculando as barras de erro
    yerr = [media - minimo, maximo - media]
    
    ax.errorbar(media.index, media, yerr=yerr, fmt='ko', capsize=5)
    ax.set_title(f"Gráfico de Erro para {j}")
    ax.set_xlabel('N', fontsize=16)
    ax.set_yscale('log')
    ax.grid(True)

axs[0].set_ylabel('Erro', fontsize=16)
plt.show()
```
